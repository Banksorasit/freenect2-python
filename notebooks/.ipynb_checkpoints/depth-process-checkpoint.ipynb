```python
%matplotlib inline

from matplotlib.pyplot import *
import numpy as np
import imageio
```

```python
points = np.load('../points.npz')['points']
undistorted = np.asarray(imageio.imread('../undistorted_depth.tiff'))
registered = np.asarray(imageio.imread('../registered_rgb.png'))
undistorted[undistorted==0.] = np.nan
undistorted = np.ma.masked_array(undistorted, mask=np.isnan(undistorted))
points = np.ma.masked_array(points, mask=np.isnan(points))
```

```python
imshow(undistorted)
```

```python
dx_dv, dx_du = np.gradient(points[..., 0])
dy_dv, dy_du = np.gradient(points[..., 1])
dz_dv, dz_du = np.gradient(points[..., 2])
```

```python
nx = dy_du*dz_dv - dz_du*dy_dv
ny = dz_du*dx_dv - dx_du*dz_dv
nz = dx_du*dy_dv - dy_du*dx_dv
normals = np.dstack((nx, ny, nz))
normals /= np.repeat(np.sqrt(np.sum(np.square(normals), axis=-1))[..., np.newaxis], 3, -1)
```

```python
figure(figsize=(14, 14))
imshow(0.5 * normals + 0.5)
axis('off')
```
